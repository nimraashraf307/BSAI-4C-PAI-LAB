### 4. Data Pre-Processing

```python
# Mean, Median, Mode

df['Age'].mean()
df['Age'].median()
df['Age'].mode()[0]
```

```python
# For Object / String Type
def fillnaObjectMode(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mode()[0])

columns = ['Name']
fillnaObjectMode(columns)
```

```python
# For Int Type
def fillnaIntMode(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mode()[0])

columns = ['Age']
fillnaIntMode(columns)
```

```python
# For Float Type
def fillnaFloatMean(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mean())

columns = ['Spa']
fillnaFloatMean(columns)
```

```python
df.info()
```

```python
df.dtypes
```

```python
def changeFloatToInt(cols):
    for i in cols:
        df[i] = df[i].astype('int64')

# Only this column should be int:
numeric_cols_to_int = ['FoodCourt']  

changeFloatToInt(numeric_cols_to_int)
```

```python
df.info()
```

```python
#Convert object columns (Label Encoding)
def convertObjToInt(cols):
    for i in cols:
        data = pd.DataFrame({i: df[i].unique()})
        data_LE = LabelEncoder()
        data_LE.fit(np.ravel(data))
        df[i] = data_LE.transform(df[i])

object_cols = ['ShoppingMall']  
convertObjToInt(object_cols)
```

```python
df.to_csv('train.csv', index=False, header=True)
```

### 5. Train - Test Splitting

```python
import pandas as pd
df = pd.read_csv("train.csv")
# df.info()
df.columns
```

```python
# Target
y = df['Spa']

# Features
X = df.drop(['PassengerId', 'Age'], axis=1)
```

```python
from sklearn.model_selection import train_test_split

train_X, test_X, train_y, test_y = train_test_split(
    X, y, test_size=0.3, random_state=42
)

print(train_X.shape, train_y.shape)
print(test_X.shape, test_y.shape)
```

### 6. Applying Classifier / Model Training

```python
categorical_cols = train_X.select_dtypes(include='object').columns
print(categorical_cols)
```

```python
print(train_y.isnull().sum())
print(test_y.isnull().sum())
```

```python
train_X = train_X[train_y.notnull()]
train_y = train_y[train_y.notnull()]

test_X = test_X[test_y.notnull()]
test_y = test_y[test_y.notnull()]
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(train_X, train_y)  
pred = model.predict(test_X)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error

print("R2 Score:", r2_score(test_y, pred))
print("MAE:", mean_absolute_error(test_y, pred))
```
